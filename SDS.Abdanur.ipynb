```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task 1: Data loading & inspection
# Load the dataset from direct URL (Titanic Passengers - Option A)
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Print first 5 rows
print("First 5 rows:")
print(df.head())

# Print last 5 rows
print("\nLast 5 rows:")
print(df.tail())

# Display column names
print("\nColumn names:")
print(df.columns.tolist())

# Display data types
print("\nData types:")
print(df.dtypes)

# Number of rows and columns
print("\nNumber of rows and columns:")
print(df.shape)

# Missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Task 2: Data types & variables
# Selected variables: Survived, Sex, Pclass, Age, Fare
# Classifications:
# - Survived: categorical (binary) – values are 0 or 1 indicating survival.
# - Sex: categorical (binary) – male or female.
# - Pclass: numerical (discrete) – integer values 1,2,3 representing classes.
# - Age: numerical (continuous) – floating point ages.
# - Fare: numerical (continuous) – floating point ticket prices.
# Justifications: Categorical for non-numeric or limited categories; numerical for measurable quantities; discrete for countable integers, continuous for real numbers.
```

```python
# Task 3: Descriptive statistics
# For Age, Fare, Pclass
numerical_vars = ['Age', 'Fare', 'Pclass']

for var in numerical_vars:
    print(f"\nStatistics for {var}:")
    print("Mean:", df[var].mean())
    print("Median:", df[var].median())
    print("Standard Deviation:", df[var].std())
    print("Variance:", df[var].var())
    print("Minimum:", df[var].min())
    print("Maximum:", df[var].max())

# Comments:
# - Age: Spread (std 14.53) shows variability; possible outliers (min 0.42, max 80); mean > median indicates slight right skew.
# - Fare: High spread (std 49.69); outliers evident (max 512.33 vs median 14.45); mean >> median shows strong right skew.
# - Pclass: Low spread (std 0.84); no real outliers as it's 1-3; mean < median due to more in class 3.
```

```python
# Task 4: Visualization
# Plot 1: Histogram of Age
plt.figure(figsize=(8,6))
df['Age'].hist(bins=20)
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Boxplot of Fare
plt.figure(figsize=(8,6))
df.boxplot(column='Fare')
plt.title('Boxplot of Ticket Fares')
plt.ylabel('Fare')
plt.show()
```

```python
# Plot 3: Scatter plot of Age vs Fare
plt.figure(figsize=(8,6))
plt.scatter(df['Age'], df['Fare'])
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
